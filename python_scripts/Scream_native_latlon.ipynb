```python
import xarray as xr
import numpy as np
```

```python
file="/work/dicad/cmip6-dev/data4freva/model/global/dyamond/DYAMOND_WINTER/LLNL/SCREAM-3km/grid.nc"
ds = xr.open_dataset(file, chunks={"time":96/2, "ncol":100})
ds_lon = ds.grid_center_lon.load()
ds_lat = ds.grid_center_lat.load()
ds_lat
```

```python
ds_lat_gt = ds_lat.where((ds_lat>-30) & (ds_lat<30),1,0)
```

```python
ds_mask = xr.Dataset({"lat_mask":xr.DataArray(ds_lat_gt.values, coords={"lat":ds_lat.values}, dims=["lat"], attrs=ds_lat_gt.attrs)}, attrs=ds_lat_gt.attrs)
ds_mask
```

```python
ds_mask.to_netcdf("/scratch/b/b380883/scream_lat_mask.nc")
```

```python
lat_gt_i = np.argwhere(ds_lat_gt.values)
```
